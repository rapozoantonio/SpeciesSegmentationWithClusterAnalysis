# Species Segmentation with Cluster Analysis

The Iris flower dataset is popular for machine learning. You can read  about it online: https://en.wikipedia.org/wiki/Iris_flower_data_set

There are 4 features: sepal length, sepal width, petal length, and petal width.

- Start with 2 clusters. Then standardize the data and do it again. 

- the Elbow rule to determine how many clusters are there.


## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

```python
df = pd.read_csv("iris-dataset.csv")
df.describe(include="all")
```

## Plot the data

Cluster the iris flowers by the shape of their sepal.

```python
x =  df["sepal_length"]
y = df["sepal_width"]
```

```python
plt.scatter(x,y)
plt.xlabel("sepal_lenght",fontsize=15)
plt.ylabel("sepal_width",fontsize=15)
plt.show()
```

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
cluster = df[["sepal_length","sepal_width"]]
```

```python
kmeans = KMeans(n_clusters=2,random_state=0).fit(cluster)
kmeans.labels_
```

```python
plt.scatter(x,y,c=kmeans.labels_,cmap="rainbow")
plt.xlabel("sepal_lenght",fontsize=15)
plt.ylabel("sepal_width",fontsize=15)
plt.show()
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data. 

```python
from sklearn import preprocessing

cluster_scaled = preprocessing.scale(df)
cluster_scaled
```

# Clustering (scaled data)

```python
kmeans = KMeans(n_clusters=2,random_state=0).fit(cluster_scaled)
kmeans.labels_
```

```python
plt.scatter(x,y,c=kmeans.labels_,cmap="rainbow")
plt.xlabel("sepal_lenght",fontsize=15)
plt.ylabel("sepal_width",fontsize=15)
plt.show()
```

Similar results as the unscaled data

## The Elbow Method

### Within Cluster Sum of Squares - WCSS

```python
wcss = []
# 'cl_num' is a that keeps track the highest number of clusters we want to use the WCSS method for. 
# We have it set at 10 right now, but it is completely arbitrary.
cl_num = 10
for i in range (1,cl_num):
    kmeans= KMeans(i)
    kmeans.fit(cluster_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
wcss
```

### The Elbow Method

```python
number_clusters = range(1,cl_num)
plt.plot(number_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

2-3 clusters should be optimal

## Clustering

```python
n_clusters= 3
kmeans = KMeans(n_clusters=n_clusters,random_state=0).fit(cluster_scaled)
kmeans.labels_
plt.scatter(x,y,c=kmeans.labels_,cmap="rainbow")
plt.xlabel("sepal_lenght",fontsize=15)
plt.ylabel("sepal_width",fontsize=15)
plt.show()
```

```python
n_clusters= 5
kmeans = KMeans(n_clusters=n_clusters,random_state=0).fit(cluster_scaled)
kmeans.labels_
plt.scatter(x,y,c=kmeans.labels_,cmap="rainbow")
plt.xlabel("sepal_lenght",fontsize=15)
plt.ylabel("sepal_width",fontsize=15)
plt.show()
```
